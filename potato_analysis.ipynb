```python
# Import Module
from color_analysis import convert_img_to_rgb

# Define file path dictionaries
fresh_potatoes = {"F_P1F": ["fresh/fresh_P1F.jpg"], 
            "F_P1B": ["fresh/fresh_P1B.jpg"], 
            "F_P2F": ["fresh/fresh_P2F.jpg"],
            "F_P2B": ["fresh/fresh_P2B.jpg"], 
            "F_P3F": ["fresh/fresh_P3F.jpg"], 
            "F_P3B": ["fresh/fresh_P3B.jpg"],
            "O_P1F": ["fresh/week_old_P1F.jpg"], 
            "O_P1B": ["fresh/week_old_P1B.jpg"],
            "O_P2F": ["fresh/week_old_P2F.jpg"], 
            "O_P2B": ["fresh/week_old_P2B.jpg"],
            "O_P3F": ["fresh/week_old_P3F.jpg"], 
            "O_P3B": ["fresh/week_old_P3B.jpg"],
            "VO_F": ["fresh/old_F.jpg"], 
            "VO_B": ["fresh/old_B.jpg"]}

aged_potatoes = {"F_P1F": ["aged/fresh_P1F.jpg"], 
            "F_P1B": ["aged/fresh_P1B.jpg"], 
            "F_P2F": ["aged/fresh_P2F.jpg"],
            "F_P2B": ["aged/fresh_P2B.jpg"], 
            "F_P3F": ["aged/fresh_P3F.jpg"], 
            "F_P3B": ["aged/fresh_P3B.jpg"],
            "O_P1F": ["aged/week_old_P1F.jpg"], 
            "O_P1B": ["aged/week_old_P1B.jpg"],
            "O_P2F": ["aged/week_old_P2F.jpg"], 
            "O_P2B": ["aged/week_old_P2B.jpg"],
            "O_P3F": ["aged/week_old_P3F.jpg"], 
            "O_P3B": ["aged/week_old_P3B.jpg"],
            "VO_F": ["aged/old_F.jpg"], 
            "VO_B": ["aged/old_B.jpg"]}

# Get rgb breakdwon of each image
for potato, link in fresh_potatoes.items():
    fresh_potatoes[potato].append(convert_img_to_rgb(link[0]))
    
for potato, link in aged_potatoes.items():
    aged_potatoes[potato].append(convert_img_to_rgb(link[0]))
```

```python
from graphing_functions import dicts_to_color_lists, display_color, get_mean_difference, display_sol_color, get_solanine_breakdown

# convert potato dicts with color breakdown to lists of each rgb color
potato_num, fresh_red_color_values, fresh_green_color_values, fresh_blue_color_values = dicts_to_color_lists(fresh_potatoes)

_, aged_red_color_values, aged_green_color_values, aged_blue_color_values = dicts_to_color_lists(aged_potatoes)

# Display bar graphs of rgb values and calculate mean difference between color values of fresh vs. aged potatoes
display_color("Red", fresh_red_color_values, aged_red_color_values, potato_num)
get_mean_difference(fresh_red_color_values, aged_red_color_values)

display_color("Green", fresh_green_color_values, aged_green_color_values, potato_num)
get_mean_difference(fresh_green_color_values, aged_green_color_values)

display_color("Blue", fresh_blue_color_values, aged_blue_color_values, potato_num)
get_mean_difference(fresh_blue_color_values, aged_blue_color_values)
```

```python
# list of all dictionary key potatoes that have solanine
solanine = ["F_P1F", "F_P1B", "O_P1F", "O_P1B"]
not_solanine = ["F_P2F", "F_P2B", "F_P3F", "F_P3B", "O_P2F", "O_P2B", "O_P3F", "O_P3B", "VO_F", "VO_B"]

# Get lists for different rgb values for aged/fresh and sol/non_sol potatoes
aged_red_sol_vals, aged_green_sol_vals, aged_blue_sol_vals, fresh_red_sol_vals, fresh_green_sol_vals, fresh_blue_sol_vals, aged_red_non_sol_vals, aged_green_non_sol_vals, aged_blue_non_sol_vals, fresh_red_non_sol_vals, fresh_green_non_sol_vals, fresh_blue_non_sol_vals = get_solanine_breakdown(solanine, not_solanine, aged_potatoes, fresh_potatoes)

# Display two graphs for each rgb value - one of fresh potatoes and one of aged
display_sol_color("Red", solanine, not_solanine, fresh_red_sol_vals, aged_red_sol_vals, fresh_red_non_sol_vals, aged_red_non_sol_vals)

display_sol_color("Green", solanine, not_solanine, fresh_green_sol_vals, aged_green_sol_vals, fresh_green_non_sol_vals, aged_green_non_sol_vals)

display_sol_color("Blue", solanine, not_solanine, fresh_blue_sol_vals, aged_blue_sol_vals, fresh_blue_non_sol_vals, aged_blue_non_sol_vals)
```
